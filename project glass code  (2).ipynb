# Machine Learning Models
# 1.DecisionTreeClassifier

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

```python
#Storing the Features into "m" variable and target column in "s" variable
m=gls.iloc[:,:-1]
s=gls[["Type"]]
```

```python
m.shape
```

```python
s.shape
```

```python
#Spliting the dataset into testing and training
mtrain,mtest,strain,stest=train_test_split(m,s,test_size=0.3,random_state=1)
dtc=DecisionTreeClassifier()
dtc.fit(mtrain,strain) # Fitting the model
```

```python
#predicting and  calculating accuracy_score
spred=dtc.predict(mtest)
accuracy=accuracy_score(stest,spred)*100
print("accuracy:",accuracy)
```

```python
#confusion_matrix
cfm=confusion_matrix(stest,spred)
cfm
```

```python
#heatmap for confusion_matrix
sns.heatmap(cfm,annot=True)
```

```python
#Classification_report
print(classification_report(stest,spred))
```

# Classification

```python
from sklearn.multiclass import OneVsRestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import roc_curve, auc
```

```python
#we are using one versus rest strategy because we are dealing with multiclass values
dtc=OneVsRestClassifier(DecisionTreeClassifier(max_features=0.2))
dtc.fit(mtrain,strain)
ypred=dtc.predict(mtest)
pred_prob=dtc.predict_proba(mtest)
```

# plot the ROC curve

```python
# Assuming 'y_Test' and 'y_pred' are multiclass
from sklearn.preprocessing import label_binarize
from sklearn.metrics import roc_curve, auc

# Binarize the output
stest_bin = label_binarize(stest, classes=np.unique(stest))
spred_bin = label_binarize(spred, classes=np.unique(stest))

# Compute ROC curve and ROC area for each class
fpr = dict()
tpr = dict()
roc_auc = dict()
n_classes = stest_bin.shape[1] # Get number of classes from binarized data
for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(stest_bin[:, i], spred_bin[:, i]) # Index binarized data
    roc_auc[i] = auc(fpr[i], tpr[i])

# Plot ROC curves for each class (example for class 0)
plt.figure()
plt.plot(fpr[0], tpr[0], color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % roc_auc[0])
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic example')
plt.legend(loc="lower right")
plt.show()
```

```python
#heatmap for confusion_matrix
sns.heatmap(cm,annot=True)
```
